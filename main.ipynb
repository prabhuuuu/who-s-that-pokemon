```python

from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.tree import DecisionTreeClassifier
```

```python
features = ['HP', 'Attack', 'Defense', 'Sp_Atk', 'Sp_Def', 'Speed', 'Generation']
X = pokemonBaseData[features]
y = pokemonBaseData['isLegendary']


from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X, y)
```

```python

X_train, X_test, Y_train, Y_test = train_test_split(X,y,test_size = 0.3, random_state = 50)

decisionTree = DecisionTreeClassifier(random_state=42)
```

```python

DTmodel = decisionTree.fit(X_train, Y_train)

DTPrediction = decisionTree.predict(X_test)
```

```python
decisionTreeAccuracy = accuracy_score(Y_test, DTPrediction) * 100

print("Accuracy(Decision Tree): %" ,decisionTreeAccuracy)

```
